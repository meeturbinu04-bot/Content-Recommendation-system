## 1. Problem Definition & Objective

Most existing content recommendation systems rely heavily on historical user interaction data such as watch history, likes, and ratings. While these approaches perform well, they often fail to capture the user's current emotional state, intent, and situational context, leading to repetitive or irrelevant recommendations.

This project aims to design a Context-Aware Content Recommendation System that generates personalized movie recommendations based on real-time user input expressed in natural language. By leveraging Large Language Models (LLMs), the system extracts user mood and intent without requiring long-term user data, making it suitable for cold-start scenarios.

The objective is to build an innovative, explainable, and ethically responsible recommendation system using pre-trained models and AI tools rather than training models from scratch.


## 4. Data Understanding and Preparation

The MovieLens dataset is used as the content source for this project. It contains metadata about movies, including titles and genres, which enables content-based filtering without requiring user interaction history.

Only the `movies.csv` file is used, as the focus of this project is on context-aware recommendation rather than collaborative filtering.


```python
import pandas as pd

# Load the MovieLens movies dataset
movies = pd.read_csv("movies.csv")

# Display basic information
movies.head()
```

```python
# Dataset shape and columns
print("Dataset Shape:", movies.shape)
print("\nColumns:", movies.columns)

# Check for missing values
movies.isnull().sum()
```

```python
# Replace missing genres if any
movies['genres'] = movies['genres'].fillna('Unknown')

# Convert genres into list format
movies['genre_list'] = movies['genres'].apply(lambda x: x.split('|'))

movies.head()
```

## 5. System Design and Architecture

The proposed system is a context-aware content recommendation system that generates movie suggestions based on the user's current mood, intent, and situational context rather than relying on historical interaction data.

User input is collected in natural language form. A Large Language Model (LLM) is used to interpret the input and extract contextual signals such as emotional state, energy level, and content intent. These signals are then mapped to relevant movie genres, which are used to filter and rank content from the dataset.

The system also provides explainable recommendations by generating human-readable reasons for each suggested movie.


### Context Variables Used

The system considers the following contextual factors:

- User Mood (e.g., happy, sad, stressed, bored)
- Energy Level (high / medium / low)
- Time-based intent (e.g., short content, relaxing content)

These variables are inferred from user-provided natural language input.


```python
# Mood to genre mapping dictionary
mood_genre_map = {
    "happy": ["Comedy", "Adventure", "Animation"],
    "sad": ["Drama", "Romance"],
    "stressed": ["Comedy", "Family"],
    "bored": ["Action", "Thriller", "Sci-Fi"],
    "relaxed": ["Fantasy", "Music"],
    "romantic": ["Romance", "Drama"],
    "angry": ["Action", "Crime"]
}
```

```python
# Simulated user input
user_input = "I had a stressful day and want something light to watch"

# Simulated extracted mood
detected_mood = "stressed"

print("User Input:", user_input)
print("Detected Mood:", detected_mood)
```

```python
# Get preferred genres for the detected mood
preferred_genres = mood_genre_map.get(detected_mood, [])

# Filter movies matching preferred genres
def genre_match(genre_list, preferred_genres):
    return any(genre in genre_list for genre in preferred_genres)

recommended_movies = movies[movies['genre_list'].apply(
    lambda x: genre_match(x, preferred_genres)
)]

recommended_movies.head()
```

```python
# Select top 5 recommendations
final_recommendations = recommended_movies.sample(5, random_state=42)

final_recommendations[['title', 'genres']]
```

### Context-Based Filtering

Based on the detected user mood, the system maps the emotional context to relevant movie genres. Movies matching these genres are filtered from the dataset, and a small set of recommendations is selected. This approach enables real-time personalization without relying on historical user data.


## 6. Core Implementation

### LLM-Based Mood and Intent Extraction

To capture the user's current emotional and contextual state, a Large Language Model (LLM) is used to analyze natural language input. The LLM extracts high-level signals such as mood and content intent, which are then mapped to relevant movie genres.

This approach enables real-time personalization and effectively addresses the cold-start problem without requiring historical user data.


### Explainable Recommendations

Modern recommendation systems often operate as black boxes, providing little insight into why specific content is suggested. To improve transparency, trust, and user understanding, this system generates human-readable explanations for each recommendation.

The explanations are aligned with the user's detected mood and intent, making the recommendation process interpretable and ethically responsible.


### Expected Output from LLM

The LLM is prompted to return a structured JSON-like response containing:
- Detected mood
- User intent

Example:
{
  "mood": "stressed",
  "intent": "light entertainment"
}


```python
def build_prompt(user_text):
    return f"""
You are an AI assistant for a movie recommendation system.
Analyze the user's input and extract:
1. The user's mood
2. The user's content intent

Respond ONLY in JSON format like this:
{{"mood": "<mood>", "intent": "<intent>"}}

User input:
"{user_text}"
"""
```

```python
# Simulated LLM response (used when API is not available)
simulated_llm_output = {
    "mood": "stressed",
    "intent": "light entertainment"
}

detected_mood = simulated_llm_output["mood"]
user_intent = simulated_llm_output["intent"]

print("Detected Mood:", detected_mood)
print("User Intent:", user_intent)
```

```python
preferred_genres = mood_genre_map.get(detected_mood, [])

print("Preferred Genres based on mood:", preferred_genres)
```

```python
# Filter movies using mood-based genres
recommended_movies = movies[movies['genre_list'].apply(
    lambda x: any(genre in x for genre in preferred_genres)
)]

# Sample final recommendations
final_recommendations = recommended_movies.sample(5, random_state=1)

final_recommendations[['title', 'genres']]
```

### AI-Assisted Personalization

By using an LLM to interpret natural language input, the system dynamically adapts recommendations to the user's current emotional and situational context. This allows the recommendation process to move beyond static history-based systems and enables more meaningful, explainable personalization.


```python
def generate_explanation(movie_title, movie_genres, mood, intent):
    return (
        f"'{movie_title}' is recommended because it belongs to the "
        f"{movie_genres} genre(s), which align well with your current mood "
        f"('{mood}') and your intent for {intent}."
    )
```

```python
# Create explainable recommendations
explainable_results = []

for _, row in final_recommendations.iterrows():
    explanation = generate_explanation(
        movie_title=row['title'],
        movie_genres=row['genres'],
        mood=detected_mood,
        intent=user_intent
    )

    explainable_results.append({
        "Title": row['title'],
        "Genres": row['genres'],
        "Explanation": explanation
    })

explainable_results
```

```python
import pandas as pd

explainable_df = pd.DataFrame(explainable_results)
explainable_df
```

### Explanation Output

Each recommended movie is accompanied by a contextual explanation that connects the user's mood and intent with the content's genre characteristics. This enhances user trust and provides clarity on how recommendations are generated.


## 7. Evaluation and Analysis

Traditional evaluation metrics such as accuracy or RMSE are not suitable for this system due to the absence of historical user ratings and model training. Instead, a qualitative, scenario-based evaluation is adopted.

The system is evaluated by testing multiple user inputs representing different moods and intents and analyzing the relevance and explainability of the recommended content.


```python
test_inputs = [
    {"text": "I feel sad and want an emotional movie", "mood": "sad", "intent": "emotional content"},
    {"text": "I'm bored and want something exciting", "mood": "bored", "intent": "high-energy entertainment"},
    {"text": "I want to relax and watch something light", "mood": "relaxed", "intent": "light entertainment"},
    {"text": "I'm feeling angry and need distraction", "mood": "angry", "intent": "intense content"}
]
```

```python
evaluation_results = []

for case in test_inputs:
    mood = case["mood"]
    intent = case["intent"]

    preferred_genres = mood_genre_map.get(mood, [])

    filtered_movies = movies[movies['genre_list'].apply(
        lambda x: any(genre in x for genre in preferred_genres)
    )]

    sampled_movies = filtered_movies.sample(3, random_state=42)

    for _, row in sampled_movies.iterrows():
        evaluation_results.append({
            "User Input": case["text"],
            "Detected Mood": mood,
            "Recommended Movie": row["title"],
            "Genres": row["genres"],
            "Intent": intent
        })
```

```python
evaluation_df = pd.DataFrame(evaluation_results)
evaluation_df
```

### Evaluation Observations

The system consistently generates recommendations aligned with the user's contextual mood and intent. Different moods result in distinct genre selections, demonstrating that the context mapping logic functions correctly. The explainable nature of the recommendations enhances interpretability and user trust.

This evaluation confirms that the system is effective for cold-start scenarios and real-time personalization.


## 8. Ethical Considerations & Responsible AI

The proposed content recommendation system is designed with ethical and responsible AI principles in mind.

### Privacy and Data Usage
The system does not collect or store sensitive personal data such as biometric information, browsing history, or location details. User input is limited to voluntary natural language text, ensuring minimal data exposure.

### Transparency and Explainability
Each recommendation is accompanied by a clear, human-readable explanation, allowing users to understand why specific content is suggested. This avoids black-box decision-making and improves user trust.

### Avoidance of Addictive Design
Unlike traditional recommendation systems that optimize for maximum screen time, this system focuses on contextual relevance rather than continuous engagement. This helps reduce content fatigue and over-consumption.

### Fairness and Bias Mitigation
The system relies on content metadata rather than demographic or identity-based attributes, minimizing the risk of bias. Users retain full control and can change inputs at any time.

Overall, the system emphasizes user well-being, transparency, and responsible use of AI technologies.
